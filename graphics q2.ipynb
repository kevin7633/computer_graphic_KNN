```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()

file_name = list(uploaded.keys())[0]

img = cv2.imread(file_name, cv2.IMREAD_GRAYSCALE)

addition = cv2.add(img, 50)

subtraction = cv2.subtract(img, 50)

multiplication = cv2.multiply(img, 1.5)

division = img.astype(np.float32) / 1.5
division = np.clip(division, 0, 255).astype(np.uint8)

titles = ['Original Image', 'Addition', 'Subtraction', 'Multiplication', 'Division']
images = [img, addition, subtraction, multiplication, division]

plt.figure(figsize=(15, 8))

for i in range(len(images)):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()

cv2.imwrite('addition_result.png', addition)
cv2.imwrite('subtraction_result.png', subtraction)
cv2.imwrite('multiplication_result.png', multiplication)
cv2.imwrite('division_result.png', division)
```
