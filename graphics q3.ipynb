```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

print("upload first image")
uploaded1 = files.upload()
file1 = list(uploaded1.keys())[0]

print("upload second image")
uploaded2 = files.upload()
file2 = list(uploaded2.keys())[0]

img1 = cv2.imread(file1, cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(file2, cv2.IMREAD_GRAYSCALE)

img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

_, binary1 = cv2.threshold(img1, 127, 255, cv2.THRESH_BINARY)
_, binary2 = cv2.threshold(img2, 127, 255, cv2.THRESH_BINARY)

and_result = cv2.bitwise_and(binary1, binary2)
or_result = cv2.bitwise_or(binary1, binary2)
not_result = cv2.bitwise_not(binary1)

titles = ['Binary Image 1', 'Binary Image 2', 'AND', 'OR', 'NOT(Image 1)']
images = [binary1, binary2, and_result, or_result, not_result]

plt.figure(figsize=(15, 8))

for i in range(len(images)):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()

cv2.imwrite('and_result.png', and_result)
cv2.imwrite('or_result.png', or_result)
cv2.imwrite('not_result.png', not_result)
```
